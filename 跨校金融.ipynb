```python
import pdfplumber
import re
import pandas as pd

file_path= "1817_109.pdf" 
pdf = pdfplumber.open(file_path) 

pattern = re.compile(r"六 ' 現金及約當現金(.*)-27 -")
file_path= "1817_109.pdf" 
pdf = pdfplumber.open(file_path) 

page = 26
p = pdf.pages[page]
text= p.extract_text()
text = text.replace("\n","")
find_text = pattern.findall(text)    
find_text[0]
```
