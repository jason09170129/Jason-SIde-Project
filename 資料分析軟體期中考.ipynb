```python
from bs4 import BeautifulSoup
import time
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
```

**爬好學校的音樂課程網頁並抓取資料:**

```python
urlbase ="https://api.hahow.in/api/group/music/courses?page="

List = []   
item_list={}  

for i in range(1,3):
  url= urlbase +str(i)
  List.append( url )

for a in List:
    res= requests.get(a,headers={"user-agent":"XXX"})
    resjson =json.loads(res.text)
    ab=res.json()['data']
    
      
    for data in ab:
      if data['status']=="PUBLISHED":
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("課程名稱:",data['title'])
        print("報名人數:",data["numSoldTickets"],"人")
        print("價錢:",data['price'],"元")
        print("評價:",data["numRating"],"則")
        print("課程大綱:","\n",data['willLearn'])

        final = {}
        item = {"People":data["numSoldTickets"],
            "Comment":data["numRating"],
            "Price":data["price"]
            
        }
        item_list[data["title"]]=item
      
  
#print(item_list)
```

**課程的DataFrame:**


```python
 df = pd.DataFrame.from_dict(item_list, orient='index')
 df
```

```python
plt.figure(figsize=(10,6))
plt.style.use("ggplot")                        #使用ggplot主題樣式
plt.xlabel("Number of applicants", fontweight = "bold",color="red",fontsize = 15)          #設定x座標標題及粗體
plt.ylabel("Comments", fontweight = "bold",color="red",fontsize = 15)                 #設定y座標標題及粗體
plt.title("Relation between Applicants and Comment",
          fontsize = 15, fontweight = "bold")          #設定標題、字大小及粗體


plt.scatter(df["People"],                            #x軸資料
        df["Comment"],                     # y軸資料
        c = "m",                       # 點顏色
        s = 60,                        # 點大小                               # 透明度
        marker = "^")                     # 點樣式
plt.show()
```

**報名人數與評價次數關係圖:可以推斷出報名人數和評價次數呈現正相關。**


```python
plt.figure(figsize=(10,6))
plt.style.use("ggplot")                             
plt.xlabel("Number of applicants", fontweight = "bold",color="red",
           fontsize = 15)                  
plt.ylabel("Price", fontweight = "bold",color="red",fontsize = 15)                 
plt.title("Relation between Applicants and Price",
          fontsize = 15, fontweight = "bold")
plt.scatter(df["People"],                            
      df["Price"],                     
      c = "g",                       
      s = 60,                                                    
      marker = "^")                     
plt.show()
```

**報名人數與價錢關係圖:看不出報名人數和價錢高低有呈現正相關。**


```python
plt.figure(figsize=(10,6))
plt.style.use("ggplot")                             
plt.xlabel("Price", fontweight = "bold",color="red",
           fontsize = 15)                  
plt.ylabel("Comments", fontweight = "bold",color="red",fontsize = 15)                 
plt.title("Relation between Comments and Price",fontsize = 15, fontweight = "bold")
plt.scatter(df["Price"],                            
      df["People"],                    
      c = "y",                       
      s = 60,                                                    
      marker = "^")                     
plt.show()
```

**價錢和評價次數的關係圖:看起來價錢和評價次數沒有相關性**

```python
plt.figure(figsize=(10,6))
df1 = df[df['Price']<=1200]#低
df2 = df[df['Price'].between(1201,2400)]#中
df3 = df[df['Price']>=2401]#高

plt.scatter("Comment", "People", data=df1,c='r', alpha = 1,label= "LOW(<=$1200)",marker="x")
plt.scatter("Comment", "People", data=df2,c='b', alpha = 0.5,label = "MID(between $1201 and $2400)")
plt.scatter("Comment", "People", data=df3,c='g', alpha = 0.5,label = "HIGH(>=$2401)",marker="^")
plt.legend()
plt.title('Comments&People',fontweight = "bold",color='red')
plt.xlabel('Comments',fontweight = "bold")
plt.ylabel('People',fontweight = "bold")


plt.show()
```
